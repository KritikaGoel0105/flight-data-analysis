```python
# Grouping the airlines and checking their mean price
data.groupby('airline')['price'].mean()
```

```python

grouped = data.groupby(['airline', 'class'])['price'].mean().unstack()

# Plot setup
grouped.plot(kind='bar', figsize=(10, 6))

# Add labels and title
plt.title("Mean Ticket Price for Each Airline", fontsize=14)
plt.xlabel("Airline", fontsize=12)
plt.ylabel("Mean Ticket Price", fontsize=12)

# Optional: rotate x labels for readability
plt.xticks(rotation=45, ha='right')

# Add legend and grid
plt.legend(title="Class")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
grouped = data.groupby('departure_time')['price'].mean()

# Plot bar chart
plt.figure(figsize=(8, 5))
plt.bar(grouped.index, grouped.values, color='skyblue', edgecolor='black')

# Add labels and title
plt.title("Mean Ticket Price by Departure Time", fontsize=14)
plt.xlabel("Departure Time", fontsize=12)
plt.ylabel("Mean Ticket Price", fontsize=12)

# Rotate x labels for readability
plt.xticks(rotation=45, ha='right')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
grouped = data.groupby('arrival_time')['price'].mean()

# Plot bar chart
plt.figure(figsize=(8, 5))
plt.bar(grouped.index, grouped.values, color='lightcoral', edgecolor='black')

# Add labels and title
plt.title("Mean Ticket Price by Arrival Time", fontsize=14)
plt.xlabel("Arrival Time", fontsize=12)
plt.ylabel("Mean Ticket Price", fontsize=12)

# Rotate x labels for better readability
plt.xticks(rotation=45, ha='right')

# Add grid lines for clarity
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
# Get unique departure times
departure_times = data['departure_time'].unique()

# Create figure with subplots
fig, axes = plt.subplots(1, len(departure_times), figsize=(5*len(departure_times), 5), sharey=True)

# Ensure axes is always iterable
axes = axes if len(departure_times) > 1 else [axes]

# Plot each departure time
for dep_time, ax in zip(departure_times, axes):
    data[data['departure_time'] == dep_time].groupby('arrival_time')['price'].mean().plot(
        ax=ax, marker='o', linestyle='-', color='teal'
    )
    ax.set_title(f"Departure Time: {dep_time}")
    ax.set_xlabel("Arrival Time")
    ax.set_ylabel("Mean Ticket Price")
    ax.tick_params(axis='x', rotation=45)
    ax.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
# Checking the Mean Ticket Price for each Source City
data.groupby('source_city')['price'].mean()
```

```python
# Checking the Mean Ticket Price for each Destination City
data.groupby('destination_city')['price'].mean()
```

```python
# Get unique source cities to create separate subplots
source_cities = data['source_city'].unique()
num_plots = len(source_cities)

# Create subplots (1 row, multiple columns)
fig, axes = plt.subplots(1, num_plots, figsize=(5*num_plots, 5), sharey=True)

# Handle the case where there’s only one source city
if num_plots == 1:
    axes = [axes]

# Plot each source city’s data
for ax, src in zip(axes, source_cities):
    subset = data[data['source_city'] == src]
    subset_grouped = subset.groupby('destination_city')['price'].mean()

    ax.plot(subset_grouped.index, subset_grouped.values, marker='o', linestyle='-', linewidth=2)
    ax.set_title(f"Source City: {src}", fontsize=12)
    ax.set_xlabel("Destination City")
    ax.set_ylabel("Mean Ticket Price")
    ax.tick_params(axis='x', rotation=45)
    ax.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
